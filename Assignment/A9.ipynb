```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
```

```python
sns.countplot(x="survived", data=titanic, hue="survived")
```

```python
titanic['sex'].value_counts().plot(kind="pie", autopct="%.2f")  
plt.show()
```

```python
plt.hist(titanic['age'], bins=5)  
plt.show()
```

```python
sns.displot(x=titanic["age"].dropna(), bins=10, kde=True)
plt.show()
```

```python
sns.boxplot(x="sex", y="age", data=titanic)
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"])
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"], hue=tips["sex"]) 
plt.show()
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"], hue=tips["sex"], 
style=tips['smoker']) 
plt.show()
```

```python
sns.barplot(x="pclass", y="age", hue="pclass", data=titanic)
plt.show()
```

```python
sns.barplot(x="pclass", y="fare", hue="sex", data=titanic)
plt.show()
```

```python
sns.boxplot(x="sex" , y="age" ,data=titanic , hue="sex")
```

```python
sns.boxplot(x="sex", y="age", hue="survived", data=titanic)
```

```python
sns.kdeplot(titanic[titanic['survived'] == 0]['age'], color="blue", label="Did not survive")
sns.kdeplot(titanic[titanic['survived'] == 1]['age'], color="orange", label="Survived")
```

```python
pd.crosstab(titanic['pclass'], titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
```

```python
sns.clustermap(pd.crosstab(titanic['parch'], titanic['survived']))
```
