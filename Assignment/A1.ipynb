```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('../Data/water_portability.csv')
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df, columns=['Potability'], prefix='', prefix_sep='',dtype=int)
df
```

```python
df_mean = df.mean() 
print(df_mean)
df = df.replace(to_replace = np.nan,value = df_mean)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
df.iloc[:,:-1] = scale.fit_transform(df.iloc[:,:-1])
df
```
